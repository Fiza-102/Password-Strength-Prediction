# TF IDF

```python
def word_divide_char(inputs):
    character=[]
    for i in inputs:
        character.append(i)
    return character    
```

```python
word_divide_char('kzde5577')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer= TfidfVectorizer(tokenizer=word_divide_char)
```

```python
X=vectorizer.fit_transform(x)   # matrix of data
```

```python
X.shape    # phle only 1 column tha but noe after vectorizer..
# column becomes 123
```

```python
# each and every feature has its importance

vectorizer.get_feature_names()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf =LogisticRegression(random_state=0,multi_class='multinomial')
```

```python
# multinomial because it had 3 classes (0,1,2)
```

```python
dt=np.array(['%@123abcd'])   # random pwd
```

```python
y_pred=clf.predict(X_test)
```
